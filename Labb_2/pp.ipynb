```python
x_values = []
y_values = []
labels = []
with open('data.txt', 'r') as file:
    # Hoppar över rad 1
    next(file)
    # Läser in och sorterar data.txt filen 
    lines = file.readlines()
    for line in lines:
        values = line.strip().split(',')
        x_values.append(float(values[0]))
        y_values.append(float(values[1]))
        labels.append(int(values[2]))
    
# Två listor för x- och y-värden baserat på label: 1 || 0
# Pikachu 1 
x_pikachu = [x_values[i] for i in range(len(labels)) if labels[i] == 1]
y_pikachu = [y_values[i] for i in range(len(labels)) if labels[i] == 1]
# Pichu 0
x_pichu = [x_values[i] for i in range(len(labels)) if labels[i] == 0]
y_pichu = [y_values[i] for i in range(len(labels)) if labels[i] == 0]

# Testpunkter
x_test = []
y_test = []
# Testpunkter
with open('testpoints.txt', 'r') as file:
    next(file)
    lines = file.readlines()

for line in lines:
    line = line[4:-2]  # Rensar testpoints.txt filen 
    parts = line.split(', ')
    x = float(parts[0])
    y = float(parts[1])
    x_test.append(x)
    y_test.append(y)
```

```python
import matplotlib.pyplot as plt
# Plotta Pichu och Pikachu
plt.scatter(x_pichu, y_pichu, marker = 'v', alpha = 0.3, color = 'orange', label = 'Pichu')
plt.scatter(x_pikachu, y_pikachu, marker = '^', alpha = 0.3, color = 'green', label = 'Pikachu')
# Plotta testpunkter
plt.scatter(x_test, y_test, s = 100,  marker = '*', color = 'red',)
# Rubriker
plt.xlabel('Width (cm)')
plt.ylabel('Height (cm)')
plt.title('Labb - 2')
plt.grid()
plt.legend(['Pichu', 'Pikachu', 'Test'])
plt.show()
```

```python
# Avståndsfunktion
import math
def distance_calc(x1, y1, x2, y2):
    # Calculate  distance
    distance = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
    return distance

# Testpunkter: avstånd från Pikachu punkter
p1_from_pikachu = [ ]
p2_from_pikachu = [ ]
p3_from_pikachu = [ ]
p4_from_pikachu = [ ]
for j in range(len(x_test)):
    for i in range(len(x_pikachu)):
        d = distance_calc(x_test[j], y_test[j], x_pikachu[i], y_pikachu[i])
        if j == 0:
            p1_from_pikachu.append(d) 
        elif j == 1:
            p2_from_pikachu.append(d)
        elif j == 2:
            p3_from_pikachu.append(d)
        elif j == 3:
            p4_from_pikachu.append(d)

# Sortera alla värden stigande
p_from_pikachu = [p1_from_pikachu, p2_from_pikachu, p3_from_pikachu, p4_from_pikachu]
for distances in p_from_pikachu:
    distances.sort()

# Testpunkter: avstånd från Pichu punkter
p1_from_pichu = [ ]
p2_from_pichu = [ ]
p3_from_pichu = [ ]
p4_from_pichu = [ ]
for j in range(len(x_test)):
    for i in range(len(x_pichu)):
        d = distance_calc(x_test[j], y_test[j], x_pichu[i], y_pichu[i])
        if j == 0:
            p1_from_pichu.append(d)
        elif j == 1:
            p2_from_pichu.append(d)
        elif j == 2:
            p3_from_pichu.append(d)
        elif j == 3:
            p4_from_pichu.append(d)
            
# Sortera alla värden från minst till störst
p_from_pichu = [p1_from_pichu, p2_from_pichu, p3_from_pichu, p4_from_pichu]
for distances in p_from_pichu:
    distances.sort()

# Klassificerar testpunkterna
for i in range(len(p_from_pikachu)):
    if p_from_pikachu[i][0] < p_from_pichu[i][0]:
        print(f'({x_test[i]}, {y_test[i]}) classified as Pikachu')
    elif p_from_pikachu[i][0] > p_from_pichu[i][0]:
        print(f'({x_test[i]}, {y_test[i]}) classified as Pichu')
```

```python
# User  input ++++++++++++++++++++++++++++++++++++++  Uppgift_1    +++++++++++++++++++++++++++++++++++++++++
# Kontrollera input
while True:
    try:
        user_x = float(input('Width: '))
        user_y = float(input('Height: '))
        if user_x < 0 or user_y < 0:
            print('Only positive numbers please!')
        else:
            break
    except ValueError:
        print('No strings only numbers please!')

# Distansen mellan input och Pikachu
user_from_pikachu = []
for i in range(len(x_pikachu)):
    d = distance_calc(user_x, user_y, x_pikachu[i], y_pikachu[i])
    user_from_pikachu.append(d)
user_from_pikachu.sort()

# Distansen mellan input och Pichu
user_from_pichu = []
for i in range(len(x_pichu)):
    d = distance_calc(user_x, user_y, x_pichu[i], y_pichu[i])
    user_from_pichu.append(d)
user_from_pichu.sort()

# Klassificerar testpunkterna
if user_from_pichu[0] < user_from_pikachu[0]:
    print(f'({user_x}, {user_y}) classified as Pichu')
elif user_from_pikachu[0] < user_from_pichu[0]:
    print(f'({user_x}, {user_y}) classified as Pikachu')
```

```python
# Tio närmaste punkter +++++++++++++++++++++++  Uppgift_2    +++++++++++++++++++++++++++++++++++++++++
my_point = [24, 34] #Valfri punkt

# Beräkna avstånden från my_point till alla Pikachu-punkter
from_all_pikachus = [distance_calc(my_point[0], my_point[1], x, y) for x, y in zip(x_pikachu, y_pikachu)]

# Beräkna avstånden från my_point till alla Pichu-punkter
from_all_pichus = [distance_calc(my_point[0], my_point[1], x, y) for x, y in zip(x_pichu, y_pichu)]

# Lägga ihop två avståndslistorna och sortera 
all_distances = from_all_pikachus + from_all_pichus
all_distances.sort()

# Slicing 10 närmaste punkterna
nearest_ten = all_distances[:10]

# Räkna Pikachu- och Pichu-punkter bland de närmaste 10 punkterna
pikachu_count = sum(1 for dist in nearest_ten if dist in from_all_pikachus)
pichu_count = sum(1 for dist in nearest_ten if dist in from_all_pichus)

# Avgör vilken klass my_point tillhör
if pikachu_count > pichu_count:
    print(my_point, 'classified as Pikachu')
elif pichu_count > pikachu_count:
    print(my_point, 'classified as Pichu')
else:
    print(f'The algorithm is not capable of classifying the point.: {my_point}')
```

```python
# ++++++++++++++++++++++++++++++++++  Uppgift  3  och 4  +++++++++++++++++++++++++++++++++++++++++

import random
accuracy = []
for x in range(10):
    # Pikach
    pikachu = list(zip(x_pikachu, y_pikachu))
    random.shuffle(pikachu)
    pikachu_50_train = pikachu[:50]
    pikachu_25_test = pikachu[50:]
    # Pichu
    pichu = list(zip(x_pichu, y_pichu))
    random.shuffle(pichu)
    pichu_50_train = pichu[:50]
    pichu_25_test = pichu[50:]
    
    test_50 = pikachu_25_test + pichu_25_test
    
    test_50_from_pikachu = []
    d_pikachu = []
    # För varje testpunk räkna avståndet till alla pikachu träningspunkter
    # För varje testpunk räkna/lagra det minsta avståndet
    for j in range(len(test_50)):
        for i in range(len(pikachu_50_train)):
            d = distance_calc(test_50[j][0], test_50[j][1], pikachu_50_train[i][0], pikachu_50_train[i][1])
            d_pikachu.append(d)
        d_pikachu.sort() 
        test_50_from_pikachu.append(d_pikachu[0]) # lagrar det minsta avståndet mellan test och pikachu punkter
        d_pikachu.clear()   
         
    # För varje testpunk räkna avståndet till alla pichu träningspunkter
    # För varje testpunk räkna/lagra det minsta avståndet
    test_50_from_pichu = []
    d_pichu = []
    for j in range(len(test_50)):
        for i in range(len(pichu_50_train)):
            d = distance_calc(test_50[j][0], test_50[j][1], pichu_50_train[i][0], pichu_50_train[i][1])
            d_pichu.append(d)
        d_pichu.sort() 
        test_50_from_pichu.append(d_pichu[0]) # lagrar det minsta avståndet mellan test och pichu punkter
        d_pichu.clear()    

    pikachu_guess = 0
    pichu_guess = 0
    # Klassificerar testpunkterna
    for i in range(50):
        if test_50_from_pikachu[i] < test_50_from_pichu[i]:
            pikachu_guess += 1
        elif test_50_from_pichu[i] < test_50_from_pikachu[i]:
            pichu_guess += 1

    final = 0 # TP + TN
    if pikachu_guess < 25:
        final = pikachu_guess + 25
    elif pichu_guess < 25:
        final = pichu_guess + 25
    else: # om båda är lika 
        final = pikachu_guess + pichu_guess
    
    accuracy.append(float(final/(pikachu_guess + pichu_guess)))

print(f'{accuracy = }')
# Medelaccuracy summan av alla items i accuracy delat på antalet items
medel_accuracy = sum(accuracy)/(len(accuracy))
print(f'medelaccuracy = {medel_accuracy:.2f}')
x_axis = list(range(1, 11))

# Plotta accuracy
plt.plot(x_axis, accuracy,)
plt.title('Accuracy')
plt.grid()
plt.show()
# Scatter
plt.scatter(x_axis, accuracy,)
plt.title('Accuracy')
plt.grid()
plt.show()
```
