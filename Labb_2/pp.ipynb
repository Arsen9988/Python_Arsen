```python
x_values = []
y_values = []
labels = []

with open('data.txt', 'r') as file:
    # börja på rad - 2
    next(file)
    lines = file.readlines()
    for line in lines:
        values = line.strip().split(',')
        x_values.append(float(values[0]))
        y_values.append(float(values[1]))
        labels.append(int(values[2]))
    



import matplotlib.pyplot as plt

# Två listor för x- och y-värden baserat på label: 1 || 0   dvs  Pikachu || Pichu
x_pikachu = [x_values[i] for i in range(len(labels)) if labels[i] == 1]
y_pikachu = [y_values[i] for i in range(len(labels)) if labels[i] == 1]



x_pichu = [x_values[i] for i in range(len(labels)) if labels[i] == 0]
y_pichu = [y_values[i] for i in range(len(labels)) if labels[i] == 0]


# Testpunkter
with open('testpoints.txt', 'r') as file:
    next(file)
    lines = file.readlines()

x_test = []
y_test = []
for line in lines:
    line = line[4:-2]  # Ta bort de första fyra tecknen från varje rad
    parts = line.split(', ')
    x = float(parts[0])
    y = float(parts[1])
    x_test.append(x)
    y_test.append(y)
    




# Plotta Pichu och Pikachu
# plt.scatter(x_pichu, y_pichu, marker = 'v', alpha = 0.3, color = 'orange', label = 'Pichu')
# plt.scatter(x_pikachu, y_pikachu, marker = '^', alpha = 0.3, color = 'green', label = 'Pikachu')
# # Plotta testpunkter
# plt.scatter(x_test, y_test, s = 100,  marker = '*', color = 'red',)

# plt.xlabel('Width (cm)')
# plt.ylabel('Height (cm)')
# plt.title('Labb - 2')
# plt.grid()

# plt.legend(['Pichu', 'Pikachu', 'Test'])
# plt.show()





# Distance calc
import math
def distance_calc(x1, y1, x2, y2):
    # Calculate  distance
    distance = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
    return distance

# Avstånd från Pikachu punkter
p1_from_pikachu = [ ]
p2_from_pikachu = [ ]
p3_from_pikachu = [ ]
p4_from_pikachu = [ ]
for j in range(len(x_test)):
    for i in range(len(x_pikachu)):
        d = distance_calc(x_test[j], y_test[j], x_pikachu[i], y_pikachu[i])
        if j == 0:
            p1_from_pikachu.append(d)
        elif j == 1:
            p2_from_pikachu.append(d)
        elif j == 2:
            p3_from_pikachu.append(d)
        elif j == 3:
            p4_from_pikachu.append(d)


# Sortera alla värden från minst till störst
p_from_pikachu = [p1_from_pikachu, p2_from_pikachu, p3_from_pikachu, p4_from_pikachu]
for distances in p_from_pikachu:
    distances.sort()



# Avstånd från Pichu punkter
p1_from_pichu = [ ]
p2_from_pichu = [ ]
p3_from_pichu = [ ]
p4_from_pichu = [ ]
for j in range(len(x_test)):
    for i in range(len(x_pichu)):
        d = distance_calc(x_test[j], y_test[j], x_pichu[i], y_pichu[i])
        if j == 0:
            p1_from_pichu.append(d)
        elif j == 1:
            p2_from_pichu.append(d)
        elif j == 2:
            p3_from_pichu.append(d)
        elif j == 3:
            p4_from_pichu.append(d)

# Sortera alla värden från minst till störst
p_from_pichu = [p1_from_pichu, p2_from_pichu, p3_from_pichu, p4_from_pichu]
for distances in p_from_pichu:
    distances.sort()


#Klassificierar testpunkterna

for i in range(len(p_from_pikachu)):
    if p_from_pikachu[i][0] < p_from_pichu[i][0]:
        print(f'({x_test[i]}, {y_test[i]}) classified as Pikachu')
    elif p_from_pikachu[i][0] > p_from_pichu[i][0]:
        print(f'({x_test[i]}, {y_test[i]}) classified as Pichu')
        

```
